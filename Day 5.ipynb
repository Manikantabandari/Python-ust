```python
# sample data representing monthly website trafic

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
traffic = [150, 2, 180, 220, 250, 310]

#create a line plot
plt.plot(months, traffic)

#add labels and a title
plt.xlabel('Month')
plt.ylabel('Monthly Traffic')
plt.title('Monthly Website Traffic')

#Display the plot
plt.show()
```

```python
#create a line plit with custom appearnce

plt.plot(months, traffic, marker='^',markersize=20, color ='green',linestyle='-.',linewidth=3)

plt.xlabel('Month')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Website Traffic')

plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_a_revenue = [45, 55, 40, 70, 80, 75]
product_b_revenue = [35, 40, 50, 55, 70, 68]

plt.plot(months, product_a_revenue, marker='o', linestyle='-', color='blue', label='Product A')

plt.plot(months, product_b_revenue, marker='s', linestyle='--', color='red', label='Product B')

plt.xlabel('March')
plt.ylabel('Monthly Revenue (in $1000)')
plt.title('Monthly Revenue Comarison')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
base_salary = 30000
hike_percentage = 10

salary = []  

for month in months:
    if month in ['Jan', 'Feb']:
        salary.append(base_salary)
    else:
        hiked_salary = base_salary + (base_salary * hike_percentage / 100)
        salary.append(hiked_salary)

plt.plot(months, salary, marker='o', linestyle='-', label='Salary After Hike')

plt.xlabel('Months')
plt.ylabel('Salary (₹)')
plt.title('Salary Hike Using Logic')
plt.legend()
plt.show()
```
