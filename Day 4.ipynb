```python
An iterator is an object that allows you to traverse through all the elemts of
a collection (like lists or tuples) one at a time

implements __iter__() and __next__() methods
```

```python
class MyNumbers:
    def __iter__(self):
      self.num = 1
      return self
    def __next__(self):
      if self.num<=5:
        x=self.num
        self.num += 1
        return x
      else:
          raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
    print(x)
```

```python
A generator is a special type of iterator that is defined using a functon
and "yeild" keyword
    uses yield instead of return
    generator values lazly
    more memory efficient than normal functions
```

```python
def my_generator():
    for i in range(1,6):
        yield i
    gen = my_generator()
    for value in gen:
        print(value)
```

```python
Iterator vs Generator

iterator: Requrires class with __iter__() and __next__()
Generator: uses function with yield keyword
iterator more code,manual implementations
Generator simple and concise implementations
iterator slightly more control
generator more memory efficient
```

```python
def numbers(n):
    for i in range(1, n + 1):
        yield i
for num in numbers(5):
    print(num)
```
