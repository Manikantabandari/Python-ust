```python
import pandas as pd
df = pd.read_csv(r"C:/Users/Administrator/Downloads/AWS Pre Assesment Test  Day  2 new.xlsx - AWS 2 result.csv")
print(df.columns)
print(df.head())
df
```

```python
score_columns = [str(i) for i in range(1, 28)]
df['Total_Score'] = df[score_columns].sum(axis=1)

max_per_question = df[score_columns].max().max()
max_possible_score = 27 * max_per_question

# Calculate percentage
df['Percentage'] = (df['Total_Score'] / max_possible_score) * 100

# Display the updated dataframe with the new columns
df[['IDs', 'Total_Score', 'Percentage']]

output_path = 'C:/Users/Administrator/Downloads/AWS Pre Assesment Test  Day 2.xlsx - AWS 2 result_with_scores.csv'
df.to_csv(output_path, index=False)
print(f"File saved successfully to: {output_path}")
print(f"Total rows saved: {len(df)}")

# Create percentage ranges and group by them
df['Percentage_Range'] = pd.cut(df['Percentage'], 
                                 bins=[0, 20, 40, 60, 80, 100], 
                                 labels=['0-20%', '20-40%', '40-60%', '60-80%', '80-100%'])

# Group by percentage range and show count and statistics
groupby_percentage = df.groupby('Percentage_Range', observed=True).agg({
    'IDs': 'count',
    'Total_Score': ['mean', 'min', 'max'],
    'Percentage': ['mean', 'min', 'max']
}).round(2)

groupby_percentage.columns = ['Count', 'Avg_Score', 'Min_Score', 'Max_Score', 'Avg_Percentage', 'Min_Percentage', 'Max_Percentage']
print("Group By Percentage Range:")
print(groupby_percentage)
```

```python
# Function to assign batch based on percentage
def assign_batch(percentage):
    if percentage < 70:
        return 'Beginner/Intermediate'
    else:
        return 'Advanced'

# Apply the function to the 'Percentage' column to create a new 'Batch' column
df['Batch'] = df['Percentage'].apply(assign_batch)

# Sort the DataFrame by 'Percentage' in descending order
df = df.sort_values(by='Percentage', ascending=False)

# Print the Name, Percentage, and Batch for all students
print(df[['IDs', 'Percentage', 'Batch']])

# Count how many students belong to each batch
batch_counts = df['Batch'].value_counts()
print("\nNumber of students in each batch:")
print(batch_counts)
```
