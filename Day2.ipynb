```python
# ASSESSMENT write the factorial of 2 using function

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
result = factorial(2)
print(result)
```

```python
# A critical operation which can raise exception is placed inside the try 
#clause & the code which handles exception is written in except clause

import sys
lst = ['b',0,2]
for entry in lst:
    try:
        print("value is:",entry)
        k = 1/int(entry)
    except:
        print("Oops", sys.exc_info(),"occured.")
        print("next entry.")
        print("***************")
print("The reciprocal of", entry,"is", k)
```

```python
# CATCHING SPECIFIC EXCEPTION
# A try clause can have any number of except clause to handle them differently
# but only one will be executed in case any exception occurs

import sys
lst = ['b',0,2]
for entry in lst:
    try:
        print("*************")
        print("Value is:",entry)
        k = 1/int(entry)
    except(ValueError):
        print("This is ValueError")
    except(ZeroDivisionError):
        print("ZeroDivisionError")
    except:
        print("Other Error")
print("The reciprocal of",entry,"is", k)
```
