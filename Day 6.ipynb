```python
#SCATTER PLOT

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.scatter(x, y, marker='o', color='blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axix')
plt.title('Scatter Plot Example')

plt.show()
```

```python
# BOX PLOT

import matplotlib.pyplot as plt

data = [12, 15, 10, 15, 17, 18, 20, 22, 23, 25, 27, 30, 35, 100, -34]
plt.boxplot(data)
plt.title('Boxplot Example with sample Data')
plt.xlabel('Data')
plt.ylabel('values')
plt.show()
```

```python
import matplotlib.pyplot as plt
quaters = ['Q1', 'Q2', 'Q3', 'Q4']
electronics = [10000, 12000, 11000, 15000]
clothing = [5000, 6000, 6500, 8000]
home_appliances = [3000, 7500, 8200, 9000]

plt.figure(figsize=(10, 6))
plt.stackplot(quaters, electronics, clothing, home_appliances, labels=['Electronics', 'Clothing', 'Home_Appliances'])

plt.xlabel('Quaters')
plt.ylabel('Sales ($)')
plt.title('Product Category Sales Over Quaters')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x)
y3 = np.tan(x)
y4 = np.exp(-x)

fig, axs = plt.subplots(2,2, figsize=(10, 8))
axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('Sine Wave')
axs[0, 1].plot(x, y2, 'b')
axs[0, 1].set_title('Cosine Wave')
axs[1, 0].plot(x, y3, 'g')
axs[1, 0].set_title('Target Wave')
axs[1, 1].plot(x, y4, 'm')
axs[1, 1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```
