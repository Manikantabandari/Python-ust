```python
lst1 = {1,2,3,4,5,6,9}
lst2 = {6,7,8,9}
print(lst1 | lst2) # union function
print(lst1.union(lst2)) #alternative command of union
print(lst2 & lst1) # intersection
print(lst1.intersection(lst2))
print(lst1 - lst2)
print(lst1.difference(lst2))
```

```python
# issubset or issuperset

print(lst2.issubset(lst1))
print(lst1.issuperset(lst2))
```
