```python
import pandas as pd

# Extract
data = pd.read_csv("C:/Users/Administrator/Downloads/Sales_data.csv")

print(data)
```

```python
# Remove null values
data = data.dropna()

# Calculate tax
data['Price_with_tax'] = data['Price'] * 1.18

print(data)
```

```python
import sqlite3

# Create database connection
conn = sqlite3.connect("sales.db")

# Load data into SQL table
data.to_sql("sales_table", conn, if_exists="replace", index=False)

print("Data loaded successfully into SQL database")

conn.close()
```
